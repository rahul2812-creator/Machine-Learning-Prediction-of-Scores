```python
import numpy as np
```

```python
no_of_skills = np.linspace(1, 40, 500, dtype='int')
```

```python
no_of_skills
```

```python
buffer = np.random.randint(1000, 30000, 500)
buffer
```

```python
salaries = (no_of_skills * 8000) + buffer
salaries
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'skills' : no_of_skills,
    'salaries' : salaries
})
```

```python
df.head(60)
```

```python
df.tail()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df.skills, df.salaries)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr
```

```python
X = df['skills'].values
Y = df['salaries'].values
```

```python
type(X), type(Y)
```

```python
X
```

```python
X = X.reshape(-1, 1)
Y = Y.reshape(-1, 1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.8)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape 
```

```python
lr.fit(X_train, Y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
lr.predict([[5]])
```

```python
X_test
```

```python
lr.predict(X_test)
```

```python
predicted_Y = lr.predict(X)
```

```python
plt.scatter(df.skills, df.salaries)
plt.plot(df.skills, predicted_Y,c='red')
```
