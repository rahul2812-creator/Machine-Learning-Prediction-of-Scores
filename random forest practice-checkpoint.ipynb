```python
import pandas as pd
```

```python
df=pd.read_csv("C:/Datasets/house-votes-84.csv")
```

```python
import numpy as np
```

```python
df.replace("?",np.nan,inplace=True)
```

```python
df.head(10)
```

```python
df.isna().sum()
```

```python
#1. Subset columns of interest
#2.Convert it into boolean according to conditions
#3.Surround with variablename[...code until step2]
```

```python
republican=df[df["Class Name"]=="republican"]
```

```python
republican.shape
```

```python
democrat=df[df["Class Name"]=="democrat"]
```

```python
democrat.shape
```

```python
for columnName in republican.columns:
    republican[columnName]=republican[columnName].fillna(republican[columnName].mode()[0])
```

```python
republican.isna().sum()
```

```python
for columnName in democrat.columns:
    democrat[columnName]=democrat[columnName].fillna(democrat[columnName].mode()[0])
```

```python
democrat.isna().sum()
```

```python
data=pd.concat([republican,democrat],axis=0)
```

```python
data.shape
```

```python
x=data.drop("Class Name",axis=1)
```

```python
x.head()
```

```python
y=data["Class Name"]
```

```python
x.isna().sum()
```

```python
y.isna().sum()
```

```python
x.dtypes
```

```python
x=pd.get_dummies(x,columns=x.columns)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42,stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
